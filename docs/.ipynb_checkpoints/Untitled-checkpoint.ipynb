```python
def ZC(C, f):
    return 1./(2*pi*1j*C*f)

def ZL(L, f):
    return 2*pi*1j*L*f
```

```python
f = linspace(9e5, 1.02e6, 1000)

plot(f, real(ZC(1e-7, f)/(ZC(3e-7, f) + ZL(8e-8, f))))
```

# Differential equation

If the potential across the capacitor is denoted by $v$, 

\begin{eqnarray}
\frac{dv(t)}{dt} &=& \frac{1}{C}i(t)
\\
\frac{di(t)}{dt} &=& -\frac{v(t)}{L} + \frac{v_{in}(t)}{L} - i(t)R
\end{eqnarray}


\begin{equation}
v_{in}(t) = LC\frac{d^2 v}{dt^2} + v
\end{equation}


\begin{equation}
\frac{d^2 v(t)}{dt^2} = \frac{v_{in}(t)-v(t)}{LC}
\end{equation}


or in the linear form

\begin{eqnarray}
\frac{dv}{dt} &=& V
\\
\frac{dV(t)}{dt} &=& -\frac{v(t)}{LC} + \frac{v_{in}(t)}{LC}
\end{eqnarray}

Let us fix the units of time in such a way that $LC = 1$. Then 

\begin{eqnarray}
\frac{dv}{dt} &=& V
\\
\frac{dV(t)}{dt} &=& -v(t)+ v_{in}(t)
\end{eqnarray}


```python
def f(V, t):
    C = 1
    L = 1
    R = .05
    f = 1
    v = V[0]
    i = V[1]
    dvdt = i/C
    didt = -v/L + sin(f*t)/L - i*R
    
    return dvdt, didt
```

```python
init = (0.0, 0.0)
t = linspace(0, 100, 10000)
res = odeint(f, init, t)
plot(t, res.T[0])
xlabel('Time')
```
